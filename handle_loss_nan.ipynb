```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

merged_df_cropped = pd.read_csv('data\wowsan\merged_df_cropped.csv')

# merged_df_cropped의 각 컬럼을 시간 축으로 그래프 그리기
# 1. 컬럼명을 리스트로 저장
columns = merged_df_cropped.columns
# Timestamp 컬럼 제거
columns = columns[1:]

# 2. 컬럼명을 순회하며 그래프 그리기
# 가로는 최대한 길게
plt.figure(figsize=(10, 25))

for column in columns:
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=merged_df_cropped, x='date', y=column)
    plt.title(column)
    plt.show()

# 3. 그래프 저장
```
