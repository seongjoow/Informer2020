## Visualization

```python
import numpy as np
```

```python
# When we finished exp.train(setting) and exp.test(setting), we will get a trained model and the results of test experiment
# The results of test experiment will be saved in ./results/{setting}/pred.npy (prediction of test dataset) and ./results/{setting}/true.npy (groundtruth of test dataset)

setting = 'informer_custom_ftMS_sl300_ll300_pl300_dm512_nh8_el2_dl1_df2048_atprob_fc5_ebtimeF_dtTrue_mxTrue_test_0(5)'

trues = np.load('./results/'+setting+'/true.npy')
preds = np.load('./results/'+setting+'/pred.npy')

# [samples, pred_len, dimensions]
trues.shape, preds.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
trues
```

```python
N, L, C = trues.shape
print("Shape of trues array: N = {}, L = {}, C = {}".format(N, L, C))
```

```python
# draw target prediction

plt.figure(figsize=(10,5))
# plt.figure()
plt.plot(trues[:,0,0], label='GroundTruth')
plt.plot(preds[:,0,8], label='Prediction') # 8 is the index of the target prediction(Throughput_0), (9 for ResponseTime_0)
plt.legend()
plt.show()
```

```python
# trues_np = trues["value"].to_numpy()
# preds_np = df_test["value"].iloc[-24*7:].to_numpy()

trues_selected = trues[0, :, 0]
preds_selected = preds[0, :, 8]

plt.figure(figsize=(20,5))
plt.plot(range(300),trues_selected, label="trues")
plt.plot(range(300),preds_selected[:], label="preds")
plt.legend()
plt.show()
```
